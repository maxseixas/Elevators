```python
import numpy as np
import matplotlib.pyplot as plt
def totaldistance(start,final):
    dist=0
    current=0
    for i in range(len(start)):
        dist+= (abs(start[i]-current) + abs(start[i]-final[i]))
        current=final[i]
    return dist

        


def swap(array, num1, num2):
    array[num1], array[num2]= array[num2], array[num1]
    return array



#current=0
#people=[]
#floors=[]
#dest=[]

#creating random intializations
#for i in range(10):
   # people.append(i)
   # fl=np.random.randint(0,99)
  #  dst=np.random.randint(0,99)
 #   while(fl==dst):
#        dst=np.random.randint(0,99)
 #   floors.append(fl)
#    dest.append(dst)



#test to make sure this is accurate
#should it be 0,9?
#print(tot, "distance")
#print("original")
#print(people)
#print(floors)
#print(dest)


#guessing and checking:

plot=[]
steps=[]

for j in range (50):
    current=0
    people=[]
    floors=[]
    dest=[]
    for l in range(10):
        people.append(l)
        fl=np.random.randint(0,99)
        dst=np.random.randint(0,99)
        while(fl==dst):
            dst=np.random.randint(0,99)
        floors.append(fl)
        dest.append(dst)

    step = 2000*j
    init=(totaldistance(floors,dest))
    for i in range(step):
        tot=totaldistance(floors,dest)
        rand1= np.random.randint(0,9)
        rand2= np.random.randint(0,9)
        while(rand1==rand2):
            rand2=np.random.randint(0,9)
        swap(people, rand1, rand2)
        swap(floors, rand1, rand2)
        swap(dest, rand1, rand2)
        newtot=totaldistance(floors,dest)
        if(newtot>tot):
            swap(people, rand1, rand2)
            swap(floors, rand1, rand2)
            swap(dest, rand1, rand2)
    
    steps.append(step)
    plot.append(init - totaldistance(floors,dest))



#print(init totaldistance(floors, dest))
#print(people)
#print(steps)
#print(plot) 
#print(people)
#print(floors)
#print(dest)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(steps, plot)
plt.xlabel('Number of Guesses')
plt.ylabel('Decrease in Distance')
plt.title('Decrease in Distance as a Function of Number of Guesses - Regular')
plt.show()
fig.savefig("RegularGraph.png", dpi = 300)


plot.sort()
print(plot[25])
```
